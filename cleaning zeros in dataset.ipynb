```python
import pandas as pd             # used to read csv file in to dataframe and perform various functions 
from pandas import ExcelWriter  # used to save cleaned dataframe as excel file.
```

```python
baseball_df = pd.read_csv("baseball_data.csv") #fetching the dataset in to pandas dataframe.
```

```python
baseball_df                     #general assessment of dataset
```

##### We found various various rows with batting average 0. This significantly proves that data for these players has not been noted. 

```python
#Assessing the data seeing no. of rows, information of data, and datatypes of various columns.
baseball_df.shape        
baseball_df.info()
baseball_df.dtypes
```

```python
#removing rows with 0 value.
baseball_df = baseball_df[baseball_df['avg']!=0.0] 
```

```python
#no of rows left after dropping unimportant one.
baseball_df.shape
```

```python
#checking for duplicates after feedback from reviewer
baseball_df.duplicated().sum()
```

```python
#checking for duplicates with exactly same name but different individuals 
baseball_df['name'].duplicated().sum()
```

```python
#Displaying duplicate names but actually different individuals
baseball_df[baseball_df['name'].duplicated(keep=False)]
```

```python
#Changing name of duplicates to Dave Roberts 1 and Dave Roberts 2 //ly for other
baseball_df.loc[206, 'name'] = 'Dave Roberts1'
baseball_df.loc[245, 'name'] = 'Dave Roberts2'
baseball_df.loc[453, 'name'] = 'Bobby Mitchell_1'
baseball_df.loc[454, 'name'] = 'Bobby Mitchell_2'
```
